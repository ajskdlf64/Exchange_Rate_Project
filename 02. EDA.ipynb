### Data Loading

```python
# library
import pandas as pd

# Warnings
import warnings
warnings.filterwarnings(action='ignore') 

# USD : 미국 달러
USD = pd.read_csv('./data/DAILY_USD.csv') ; USD = USD.set_index('Unnamed: 0')

# EUR : 유로
EUR = pd.read_csv('./data/DAILY_EUR.csv') ; EUR = EUR.set_index('Unnamed: 0')

# JPY(100) : 일본 옌
JPY = pd.read_csv('./data/DAILY_JPY(100).csv') ; JPY = JPY.set_index('Unnamed: 0')

# AED : 아랍에미리트 디르함
AED = pd.read_csv('./data/DAILY_AED.csv') ; AED = AED.set_index('Unnamed: 0')

# AUD : 호주 달러
AUD = pd.read_csv('./data/DAILY_AUD.csv') ; AUD = AUD.set_index('Unnamed: 0')

# BHD : 바레인 디나르
BHD = pd.read_csv('./data/DAILY_BHD.csv') ; BHD = BHD.set_index('Unnamed: 0')

# BND : 브루나이 달러
BND = pd.read_csv('./data/DAILY_BND.csv') ; BND = BND.set_index('Unnamed: 0')

# CAD : 캐나다 달러
CAD = pd.read_csv('./data/DAILY_CAD.csv') ; CAD = CAD.set_index('Unnamed: 0')

# CHF : 스위스 프랑
CHF = pd.read_csv('./data/DAILY_CHF.csv') ; CHF = CHF.set_index('Unnamed: 0')

# CNH : 위안화
CNH = pd.read_csv('./data/DAILY_CNH.csv') ; CNH = CNH.set_index('Unnamed: 0')

# DKK : 덴마아크 크로네
DKK = pd.read_csv('./data/DAILY_DKK.csv') ; DKK = DKK.set_index('Unnamed: 0')

# GBP : 영국 파운드
GBP = pd.read_csv('./data/DAILY_GBP.csv') ; GBP = GBP.set_index('Unnamed: 0')

# HKD : 홍콩 달러
HKD = pd.read_csv('./data/DAILY_HKD.csv') ; HKD = HKD.set_index('Unnamed: 0')

# IDR(100) : 인도네시아 루피아
IDR = pd.read_csv('./data/DAILY_IDR(100).csv') ; IDR = IDR.set_index('Unnamed: 0')

# KWD : 쿠웨이트 디나르
KWD = pd.read_csv('./data/DAILY_KWD.csv') ; KWD = KWD.set_index('Unnamed: 0')

# MYR : 말레이지아 링기트
MYR = pd.read_csv('./data/DAILY_MYR.csv') ; MYR = MYR.set_index('Unnamed: 0')

# NOK : 노르웨이 크로네
NOK = pd.read_csv('./data/DAILY_NOK.csv') ; NOK = NOK.set_index('Unnamed: 0')

# NZD : 뉴질랜드 달러
NZD = pd.read_csv('./data/DAILY_NZD.csv') ; NZD = NZD.set_index('Unnamed: 0')

# SAR : 사우디 리얄
SAR = pd.read_csv('./data/DAILY_SAR.csv') ; SAR = SAR.set_index('Unnamed: 0')

# SEK : 스웨덴 크로나
SEK = pd.read_csv('./data/DAILY_SEK.csv') ; SEK = SEK.set_index('Unnamed: 0')

# SGD : 싱가포르 달러
SGD = pd.read_csv('./data/DAILY_SGD.csv') ; SGD = SGD.set_index('Unnamed: 0')
 
# THB : 태국 바트
THB = pd.read_csv('./data/DAILY_THB.csv') ; THB = THB.set_index('Unnamed: 0')

# Concat
DAILY_DATA = pd.concat([AED,AUD,BHD,BND,CAD,CHF,CNH,DKK,EUR,GBP,HKD,IDR,\
                        JPY,KWD,MYR,NOK,NZD,SAR,SEK,SGD,THB,USD] ,axis=1)

# Index Setting
DAILY_DATA.index = DAILY_DATA.index.rename("date")


# Data Save
DAILY_DATA.to_csv('./data/DAILY_DATA.csv')
```

### USD Correlation

```python
# USD Correlation
usd_corr = DAILY_DATA.corr()[-1:].T
usd_corr['COUNTRY'] = usd_corr.index

# Seaborn Bar Plot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,6))
sns.set(style="whitegrid")
ax = sns.barplot(x = 'COUNTRY', y = 'USD', data = usd_corr)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

### Distribution

```python
# Country
COUNTRY = 'USD'

# Distribution Plot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (18,6))
plt.title('{} Price Distribution 2010 ~ 2019'.format(COUNTRY), fontsize = 30, position = (0.5, 1.0+0.05))
sns.distplot(DAILY_DATA[COUNTRY], kde = True, color = 'b')
plt.show()
```
